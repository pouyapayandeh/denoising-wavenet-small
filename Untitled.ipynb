```python

from __future__ import division
from main import load_config
import os
import util
import models
import os
import util
import tqdm
import numpy as np
import scipy
%matplotlib notebook
import matplotlib.pyplot as plt
from IPython.display import Audio
```

```python
noisy_input_path = 'data/NSDTSEA/noisy_testset_wav/'
config = load_config("sessions/003/config.json")
```

```python
if noisy_input_path.endswith('.wav'):
    filenames = [noisy_input_path.rsplit('/', 1)[-1]]
    noisy_input_path = noisy_input_path.rsplit('/', 1)[0] + '/'
#     if cla.clean_input_path is not None:
#         cla.clean_input_path = cla.clean_input_path.rsplit('/', 1)[0] + '/'
else:
    if not noisy_input_path.endswith('/'):
        noisy_input_path += '/'
    filenames = [filename for filename in os.listdir(noisy_input_path) if filename.endswith('.wav')]

filename = filenames[0]
noisy_input = util.load_wav(noisy_input_path + filename, config['dataset']['sample_rate'])
input = {'noisy': noisy_input, 'clean': None}
output_filename_prefix = filename[0:-4] + '_'
```

```python
model = models.Pnet5(config, target_field_length=None,load_checkpoint=None, print_model_summary=False)
```

```python
batch_size =1
if len(input['noisy']) < model.receptive_field_length:
    raise ValueError('Input is not long enough to be used with this model.')

num_output_samples = input['noisy'].shape[0] - (model.receptive_field_length - 1)
num_fragments = int(np.ceil(num_output_samples / model.target_field_length))
num_batches = int(np.ceil(num_fragments / batch_size))

denoised_output = []
noise_output = []
num_pad_values = 0
fragment_i = 0
layer_outputs = None
input_batches = []
for batch_i in tqdm.tqdm(range(0, num_batches)):

    if batch_i == num_batches-1: #If its the last batch'
        batch_size = num_fragments - batch_i*batch_size

    input_batch = np.zeros((batch_size, model.input_length))

    if fragment_i + model.target_field_length > num_output_samples:
        remainder = input['noisy'][fragment_i:]
        current_fragment = np.zeros((model.input_length,))
        current_fragment[:remainder.shape[0]] = remainder
        num_pad_values = model.input_length - remainder.shape[0]
    else:
        current_fragment = input['noisy'][fragment_i:fragment_i + model.input_length]

    input_batch[0, :] = current_fragment
    fragment_i += model.target_field_length

#     denoised_output_fragments = model.denoise_batch({'data_input': input_batch})
    input_batches.append(input_batch)
```
